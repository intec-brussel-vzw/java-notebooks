# Overzicht

Aan het einde van de cursus zou u de volgende doelstellingen moeten kunnen bereiken:  
• Beschrijf de voordelen van Spring Boot  
• Initialiseer een project met Spring Boot Starters  
• Maak gebruik van de automatische configuratiefuncties  
van Spring Boot • Maak vereenvoudigde backing-store-oplossingen met Spring Data JPA  
• Bouw een eenvoudige MVC-applicatie met Spring Boot, embedded webserver en dikke JARs of klassieke WAV's  
• Bouw een RESTful webapplicatie  
• Gebruik Spring Boot-verbeteringen om te testen  
• Gebruik Spring Security om web- en REST-eindpunten  
te beveiligen • Schakel metrische gegevens en bewakingsmogelijkheden in en uit met behulp van Spring Boot-actuator  
• Maak gebruik van geavanceerde configuratiemogelijkheden

## Voorwaarden

• Een goede praktische kennis van webapplicatie ontwikkeling met behulp van Java en een IDE (Eclipse, STS of  
IntelliJ).  
• Basiskennis van Spring: Java Config, component-scanning, Spring driven testing en Spring data-management  
(JdbcTemplate, @Transactional).  
• Ervaring met het gebruik van Java en het bouwen van tools zoals Maven of Gradle.
