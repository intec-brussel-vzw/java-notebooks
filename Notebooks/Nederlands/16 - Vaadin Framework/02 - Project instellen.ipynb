# Tutorial: import, run and debug a Maven Spring Boot project in IntelliJ

Learn how to open and debug a Vaadin project in IntelliJ. Set up IntelliJ for better autocomplete support.


## [](https://vaadin.com/docs/latest/tutorial/project-setup/#downloading-a-vaadin-application-starter)Een Vaadin Application Starter downloaden

Deze tutorial maakt gebruik van een vooraf geconfigureerde starter van Vaadin Start. De startersapplicatie omvat:

-   Een gegevensmodel dat bestaat uit , en JPA entiteiten.`Contact``Company``Status`
    
-   Spring Data-opslagplaatsen voor het aanhouden en ophalen van de entiteiten uit een ingesloten H2-database.
    
-   Een gegevensgenerator die de database vult met testgegevens.
    
-   Een enkele, lege weergave.
    
-   Een Dockerfile.

## [](https://vaadin.com/docs/latest/tutorial/project-setup/#importing-a-maven-project-into-intellij)Een Maven-project importeren in IntelliJ

1.  Pak het gedownloade archief uit op een bestandslocatie naar keuze. Vermijd het uitpakken naar de downloadmap, omdat u uw project onbedoeld kunt verwijderen bij het wissen van oude downloads.
    
2.  Selecteer in IntelliJ **openen** in het welkomstscherm of **het menu Bestand**.
    
    ![IntelliJ-importscherm](https://vaadin.com/docs/latest/static/548eaf47faff0a28692175fd167432cf/03979/intellij-import.png)
    
3.  Zoek de uitgepakte map en selecteer het bestand.`pom.xml`
    
    ![open pom file](https://vaadin.com/docs/latest/static/eb490cbbf4d8fdd1fed9575b153016a3/03979/open-pom.png)
    
4.  Select **Open as Project**. This imports a project based on the POM file.
    
    ![open as project](https://vaadin.com/docs/latest/static/ff46a2fbd359543a90bcd6fa619acd3b/03979/open-as-project.png)
    
5.  IntelliJ imports the project and downloads all necessary dependencies. This can take several minutes, depending on your internet connection speed.
    

When the import is complete, your project structure looks like this:

-   Java source files are in the folder.`src/main/java`
    

![project file hierarchy](https://vaadin.com/docs/latest/static/491c3f4e303b4b06615755c25b62e0c0/85ae3/project-structure.png)

Spring Boot makes it easier to run a Java web application, because it takes care of starting and configuring the server.

To run your application, run the Application class that contains the method that starts Spring Boot. IntelliJ automatically detects that you have a class with a method and displays it in the **run configurations** dropdown.`main()``main()`

To start your application:

-   Open and click the play button next to the code line containing the method.`Application.java``main()`
    
-   After you have run the application once from the method, it shows up in the **run configurations** dropdown in the main toolbar. On subsequent runs, you can run the application from there.`main()`
    

![run button locations](https://vaadin.com/docs/latest/static/14d61b41a95d6387eb9800934c630f0d/03979/run-app.png)

The first time you start a Vaadin application, it downloads frontend dependencies and builds a JavaScript bundle.

## [](https://vaadin.com/docs/latest/tutorial/project-setup/#enabling-auto-import-in-intellij)Enabling Auto Import in IntelliJ

You can configure IntelliJ to automatically resolve imports for Java classes. This makes it easier to copy code from this tutorial into your IDE.

To enable auto import in IntelliJ:

1.  Open the **Preferences/Settings** window and navigate to **Editor > General > Auto Import**.
    
2.  Enable the following two options:
    
    -   **Add unambiguous imports on the fly**.
        
    -   **Optimize imports on the fly**.
        
        ![automatic import settings](https://vaadin.com/docs/latest/static/8d4cf0a1336bfa7a5a7222f5a9599ed0/03979/auto-import.png)
        
        Vaadin shares many class names (like Button) with Swing, AWT, and JavaFX.
        
    
3.  If you don’t use Swing, AWT, or JavaFX in other projects, add the following packages to the **Exclude from import and completion** list to help IntelliJ select the correct classes automatically.
    
    -   `com.sun`
        
    -   `java.awt`
        
    -   `javafx.scene`
        
    -   `javax.swing`
        
    -   `jdk.internal`
        
    -   `sun.plugin`
        
    

Now that you have a working development environment, you are ready to start building a web application.
